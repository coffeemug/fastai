```python
# Disable logging so output isn't flooded by nonsense
import logging
logging.disable(logging.WARNING)

# Log into huggingface
from pathlib import Path
from huggingface_hub import notebook_login

if not (Path.home()/'.cache/huggingface'/'token').exists():
  notebook_login()
```

```python
# Create a pipeline
import torch
from diffusers import StableDiffusionPipeline
pipe = StableDiffusionPipeline.from_pretrained(
    "CompVis/stable-diffusion-v1-4",
    revision="fp16",
    torch_dtype=torch.float16
).to("cuda")
```

```python
# let's go!
prompt="painting of a beautiful iranian woman, black curly hair, rembrant style"
pipe(prompt, negative_prompt="head cover, frame, masculine").images[0]
```
